```python
import matplotlib.pyplot as plt
import numpy as np

# Месяцы
months = np.arange(1, 13)
month_names = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
               'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']

# Реальные подписчики (тысяч человек)
actual = np.array([5.2, 6.8, 9.5, 11.2, 13.8, 17.5, 22.3, 24.1, 27.8, 31.2, 33.5, 34.0])

# Прогноз на год (сделаный в начале года)
forecast = np.array([5.0, 7.5, 8.5, 12.5, 14.5, 16.0, 23.5, 25.5, 26.5, 30.0, 34.5, 35.5])

# Промежуточные цели (тысяч подписчиков)
milestones = [10, 20, 30, 40]
milestone_names = ['Бронза', 'Серебро', 'Золото', 'Платина']
```

```python
plt.plot(months, actual, color='blue', label='Факт', marker='o')
plt.plot(months, forecast, color='red', label='Прогноз', marker='s', linestyle='--')
plt.title('Рост подписчиков Youtube-каналов в 2026 году')
plt.xlabel('Месяц')
plt.ylabel('Подписчиков (тыс.)')
plt.xticks(months, month_names, rotation=45)
plt.axhline(10, color='brown', linestyle='--')
plt.text(x=13, y=10, s='Бронза', color='brown')

plt.axhline(20, color='silver', linestyle='--')
plt.text(x=13, y=20, s='Серебро', color='silver')

plt.axhline(30, color='gold', linestyle='--')
plt.text(x=13, y=30, s='Золото', color='gold')

plt.axhline(40, color='aqua', linestyle='--')
plt.text(x=13, y=40, s='Платина', color='aqua')

plt.fill_between(months, actual, forecast, color='red', where=actual<forecast, interpolate=True, alpha=0.2, label='Факт ниже прогноза')
plt.fill_between(months, actual, forecast, where=actual>forecast, color='green', alpha=0.2, interpolate=True, label="Факт выше прогноза")
plt.legend()
plt.show()
```
