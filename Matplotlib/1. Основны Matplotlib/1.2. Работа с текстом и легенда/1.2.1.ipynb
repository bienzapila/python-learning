```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

x = np.arange(26)
y = np.random.randint(500, 10000, 26)

plt.plot(x, y)
plt.title("Хуй залупа вертолет",
          loc = 'right',
          fontsize = '14',
          color = 'green',
          fontname = 'times new roman',
          style = 'italic')
plt.xlabel('Говно')
plt.ylabel('Залупа')
plt.annotate('Хуй', xy= (5, 7000), xytext=(10, 7000))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

x = np.arange(26)
y = np.random.randint(500, 10000, 26)

plt.plot(x, y)
plt.title("Выручка магазина: первое полугодие 2026",
          loc='center',
          fontsize=16,
          color='navy',
          fontweight='bold',
          style='italic',
          fontname='Arial')

plt.xlabel('Недели 2026 года')
plt.ylabel('Выручка (тыс. руб.)')
plt.annotate('Людям надоел наш продукт',
             xy=(5, 6800),
             xytext=(3, 9000),
             fontsize=12,
             color='red',
             fontweight='bold',

             arrowprops=dict(        # Настройки стрелки
                 facecolor='black',  # цвет заливки стрелки
                 edgecolor='gray',   # цвет контура стрелки
                 linewidth=0.5,      # толщина линии стрелки
                 shrink=0.1          # отступ от точки и текста (10%)
             ))
plt.show()
```

```python
import random as rnd
plt.plot([rnd.randint(1, 101) for _ in range(15)], label='Линия')
plt.title('Рандомные числа', fontname='times new roman')
plt.xlabel('Номер', fontname = 'times new roman')
plt.ylabel('Значение', fontname = 'times new roman')
plt.text(7, 90, 'Вот такой крутой график')
plt.legend(loc = 'best')
plt.show()
```
