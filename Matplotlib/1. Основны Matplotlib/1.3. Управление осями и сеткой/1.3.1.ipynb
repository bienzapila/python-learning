```python

months = np.arange(1, 13)
month_names = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
               'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']

# Выручка по категориям (тыс. руб.)
electronics = [450, 420, 480, 510, 620, 700, 680, 720, 690, 650, 800, 1200]
clothing = [380, 350, 420, 450, 520, 550, 600, 620, 580, 540, 700, 850]
home = [220, 200, 240, 260, 300, 320, 340, 350, 330, 310, 400, 550]

plt.plot(months, electronics, label='Электроника')
plt.plot(months, clothing, label='Одежда')
plt.plot(months, home, label='Товары для дома')

plt.title("Динамика продаж по категориям в 2023 году")
plt.xlabel("Месяц")
plt.ylabel("Выручка (тыс. руб.)")
plt.legend()

plt.xlim(0.5, 12.5)
plt.ylim(0, 1400)

plt.xticks(months, month_names, rotation=45)

plt.grid(True,
         linestyle='--',
         color='gray',
         alpha=0.3,
         linewidth=0.8,
         axis='both')

# Устанавливаем частоту линий по y
plt.yticks(np.arange(0, 1401, 200))

plt.show()
```
